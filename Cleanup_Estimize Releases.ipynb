# Tricky Prep - Isolating and Creating a Feature out EPS and REV

```python
#Now we have to clean up EPS and REV. They are each in the dataset, and we want to makes features out of them

feature_names=('fiscal_year','fiscal_quarter','reports_at','instrument_id',
               'ticker','Y','sector','industry',)
#NOTE: Putting in Y, removes all future data observations

UAA_df1 = UAA_df.pivot_table(index=feature_names,
                     columns='metric', 
                     values='estimize_mean',)
UAA_df1 = UAA_df1.reset_index().rename_axis(None, axis=1)
UAA_df1.rename(columns={'eps': 'EMZ_eps', 'revenue':'EMZ_rev'}, inplace=True)

#UAA_df1.info()
UAA_df1.tail()

#Colums that have been remeoved from dataset: 'metric', 'actual', 'metric_id','actual'
#print(UAA_df1)
#print(feature_names)
```
